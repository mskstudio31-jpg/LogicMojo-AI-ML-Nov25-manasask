### take positive interger and print number of digits

```python
n = int (input("Enter positive integer: "))
if n<0:
    print("Please enter positive integer")
else:
    print(f"entered number = {n}")
    num_digits = 0
    while n > 0:
        n=n // 10
        num_digits += 1
    print(f"there are {num_digits} number of digits")
       

     
   
    
```

### nested loops

```python
for i in range(1,6):
    print(f"The multiplication of table of {i}\n")
    for j in range(1, 11):
        print(f"{i} *{j} = {i*j}")
    print("\n.......................")
                 
    
          
```
