sorted function return sorted list for specified iterable object
sorted(iterable, key, reverse)

-iterable - required argument. This is iterable object to sort
key - optional argument. defauly is none. Key is decide the order
-reverse - optional boolean argument. default false

```python
a=["app", "gaa", "bba", "paa", "caa", "ff" ]
```

```python
sort_a =sorted(a)
sort_a
```

```python
city= ['Newyork', 'paris', 'london', 'sydney', 'tokyo']
sorted(city)
```

```python
sorted(city, key=len)

sorted(city, reverse=True)
```

```python
my_list  = [[1,2,3], [4,5,6], [-5, 6,9], [1,1,1]]
sorted(my_list, key=sum)
```

keywords

```python
def func(**kwargs):
    for k,v in kwargs.items():
        print(f"{k}:{v}")
```

```python
def add_num(*args):
    result=0
    for a in args:
        result += a
    return result
    

```

```python
add_num(1,4,6,8)
```
